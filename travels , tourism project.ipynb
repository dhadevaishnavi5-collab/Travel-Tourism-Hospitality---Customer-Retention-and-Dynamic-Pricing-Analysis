Exploratory Data Analysis EDA Analysis 

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True
)

plt.show()
```

monthly Tourism Revenue 

```python
import matplotlib.pyplot as plt

monthly_revenue = df.groupby('Month')['Tourism_Revenue_USD'].sum()

monthly_revenue.plot(kind='line')
plt.title('Monthly Tourism Revenue')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.show()
```

```python
sns.boxplot(x=df['Average_Room_Price_USD'])
plt.show()
```

```python
monthly_tourists = df.groupby('Month')['Number_of_Tourists'].sum()

monthly_tourists.plot(kind='line', marker='o')
plt.title('Monthly Tourist Arrivals')
plt.xlabel('Month')
plt.ylabel('Number of Tourists')
plt.show()
```

Revenue Pattern 

```python
monthly_revenue = df.groupby('Month')['Tourism_Revenue_USD'].sum()

monthly_revenue.plot(kind='bar')
plt.title('Monthly Tourism Revenue')
plt.xlabel('Month')
plt.ylabel('Revenue (USD)')
plt.show()
```

Country Wise Revenue


```python
country_revenue = df.groupby('Country')['Tourism_Revenue_USD'].sum()

country_revenue.sort_values(ascending=False).head(10)
```

Customers monthly Hotel Accupancy Rate

```python
occupancy = df.groupby('Month')['Hotel_Occupancy_Rate'].mean()

occupancy.plot(kind='line', marker='o')
plt.title('Monthly Hotel Occupancy Rate')
plt.xlabel('Month')
plt.ylabel('Occupancy Rate (%)')
plt.show()
```

Room price accuracy
